## AST

```python
# how ast works
```

```python
import ast
```

```python
code = """
import math

def square(n):
    return n * n

def check_number(n):
    if n > 0:
        print("Positive")
    else:
        print("Negative")

for i in range(3):
    check_number(i)
"""
```

```python
tree = ast.parse(code)
```

```python
print("----- AST Structure -----")
print(ast.dump(tree, indent=4))
```

```python
print("\n----- Using ast.walk() -----")
for node in ast.walk(tree):
    if isinstance(node, ast.FunctionDef):
        print("Function found:", node.name)
    elif isinstance(node, ast.If):
        print("If statement found")
    elif isinstance(node, ast.For):
        print("For loop found")
    elif isinstance(node, ast.Call):
        print("Function call found")
    elif isinstance(node, ast.Assign):
        print("Assignment found")
```

```python
print("\n----- Using NodeVisitor -----")

class MyVisitor(ast.NodeVisitor):
    def visit_FunctionDef(self, node):
        print("Visited function:", node.name)
        self.generic_visit(node)  # visit children
    def visit_If(self, node):
        print("Visited an if statement")
        self.generic_visit(node)
    def visit_For(self, node):
        print("Visited a for loop")
        self.generic_visit(node)

visitor = MyVisitor()
visitor.visit(tree)
```

```python
print("\n----- Using NodeTransformer -----")

class MyTransformer(ast.NodeTransformer):
    def visit_FunctionDef(self, node):
        node.name = node.name + "_new"
        return self.generic_visit(node)

transformer = MyTransformer()
new_tree = transformer.visit(tree)

# Dump new tree to see changes
print(ast.dump(new_tree, indent=4))
```

## Text preprocessing

```python
df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmitizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):

  #lowercasing
  text = text.lower()

  #removing punctuatuions
  text = re.sub(r'[^\w\s]', '', text)
  words = text.split()

  # stop word removal
  words = [word for word in words if word not in stop_words]

  #lemmitize
  words = [lemmitizer.lemmatize(word) for word in words]

  return " ".join(words)

```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

## Text Vectorization

```python
df = df.reset_index(drop= True)
```

```python
indices = pd.Series(df.index , index = df['title'])
```

```python
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer( max_features=50000 , ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix
```

## cosine similarity and Model Building

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title , n = 10):

  if title not in indices:
    return "Movie not Found"

  idx = indices[title]

  sim_score = cosine_similarity(tfidf_matrix[idx] , tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]

  return df['title'].iloc[similar_idx]
```

```python
recommend('Toy Story')
```
